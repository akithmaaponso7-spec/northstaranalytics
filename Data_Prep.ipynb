```python
import pandas as pd
import numpy as np
import os
```

```python
folder_path = "northstar_dataset/northstar_dataset"

customers = pd.read_csv(f"{folder_path}/customers.csv")
orders = pd.read_csv(f"{folder_path}/orders.csv")
deliveries = pd.read_csv(f"{folder_path}/deliveries.csv")
drivers = pd.read_csv(f"{folder_path}/drivers.csv")
vehicles = pd.read_csv(f"{folder_path}/vehicles.csv")
hubs = pd.read_csv(f"{folder_path}/hubs.csv")
complaints = pd.read_csv(f"{folder_path}/complaints.csv")
incidents = pd.read_csv(f"{folder_path}/incidents.csv")
app_events = pd.read_csv(f"{folder_path}/app_events.csv")
data_dictionary = pd.read_csv(f"{folder_path}/data_dictionary.csv")
```

```python
datasets = {
    "customers": customers,
    "orders": orders,
    "deliveries": deliveries,
    "drivers": drivers,
    "vehicles": vehicles,
    "hubs": hubs,
    "complaints": complaints,
    "incidents": incidents,
    "app_events": app_events
}

for name, df in datasets.items():
    print(name, df.shape)
```

```python
for name, df in datasets.items():
    print("\n", name)
    print(df.info())
```

```python
for name, df in datasets.items():
    print("\nMissing values in", name)
    print(df.isnull().sum())
```

```python
for name, df in datasets.items():
    print(name, "duplicates:", df.duplicated().sum())
```

```python
for name, df in datasets.items():
    df.drop_duplicates(inplace=True)
```

```python
for df in datasets.values():
    for col in df.columns:
        if "date" in col.lower() or "time" in col.lower():
            df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
for name, df in datasets.items():
    df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
if "scheduled_delivery_time" in deliveries.columns and "actual_delivery_time" in deliveries.columns:
    deliveries["delay_minutes"] = (
        deliveries["actual_delivery_time"] - deliveries["scheduled_delivery_time"]
    ).dt.total_seconds() / 60
```

```python
if "delay_minutes" in deliveries.columns:
    deliveries["delivery_status"] = np.where(
        deliveries["delay_minutes"] > 0,
        "Delayed",
        "On Time"
    )
```

```python
customers.to_csv("cleaned_customers.csv", index=False)
orders.to_csv("cleaned_orders.csv", index=False)
deliveries.to_csv("cleaned_deliveries.csv", index=False)
drivers.to_csv("cleaned_drivers.csv", index=False)
vehicles.to_csv("cleaned_vehicles.csv", index=False)
hubs.to_csv("cleaned_hubs.csv", index=False)
complaints.to_csv("cleaned_complaints.csv", index=False)
incidents.to_csv("cleaned_incidents.csv", index=False)
app_events.to_csv("cleaned_app_events.csv", index=False)
```
